Saya punya sekantong permen. Terdapat lima jenis
permen. Buat diagram yang menunjukkan
persentase peluang bahwa kita akan
mengeluarkan permen Kopiko dari kantong jika
kita melakukan pengambilan acak.
Berapa peluang memilih permen Kopiko?

nama_permen = [‘Mentos', ‘Kopiko', ‘Golia', ‘Yupie’, ‘Fisherman’]
Jumlah_permen = [52, 39, 78, 13, 78]


```python
nama_permen = ['Mentos', 'Kopiko', 'Golia', 'Yupie', 'Fisherman']
jumlah_permen = [52, 39, 78, 13, 78]

warna = ("#d8eded", "#141c14", "#004fa8", "#038005", "#a2bbf5")
target = (0, 0.2, 0, 0, 0)
plt.figure(figsize=(15, 15))
plt.title("Distribusi Product", y=1.05)

plt.pie(
    jumlah_permen,
    labels=nama_permen,
    autopct='%1.2f%%',
    colors=warna,
    explode=target,
    shadow=True,
)
plt.show()

#RUMUS PELUANG
total_permen = sum(jumlah_permen)
index_kopiko = nama_permen.index("Kopiko")
peluang_kopiko = jumlah_permen[index_kopiko]/total_permen


print("Peluang dari terambilnya kopiko adalah", peluang_kopiko)

```

**Jadi, dapat disimpulkan bahwa peluang dari terambilnya kopiko pada pengambilan secara acak adalah 0.15**
